```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import numpy as np
import pandas as pd
```

```python
dataSet = pd.read_csv('mushrooms.data', header = None, nrows=5000)
dataSet = dataSet.apply(lambda x: pd.factorize(x)[0])
targetSet = dataSet.copy().drop(columns=range(1,23))
dataSet = dataSet.drop(columns=0)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(dataSet, targetSet, test_size=.3)
```

```python

cls = GaussianNB(var_smoothing = 0.000009)
ytrainFit = np.ravel(ytrain)
cls.fit(xtrain, ytrainFit)
```

```python
ypred = cls.predict(xtest)
print(classification_report(ytest, ypred))
```

```python
print('\nTest Accuracy : {:.3f}'.format(cls.score(xtest, ytest))) 
print('Training Accuracy : {:.3f}'.format(cls.score(xtrain, ytrainFit)))
```

```python
print('Подбор гиперпараметра var_smoothing:')
i = 0
res = pd.DataFrame([], columns=["var_sm", "train%", "test%"])
for smoothing in np.arange(0.97, 0.6, -0.0009):
    cl = GaussianNB(var_smoothing=smoothing)
    cl.fit(xtrain, ytrainFit)
    res.loc[i] = [round(smoothing, 10), cl.score(xtrain, ytrainFit)*100, cl.score(xtest, ytest)*100]
    i+=1

print(res.to_markdown())
res
    
```
